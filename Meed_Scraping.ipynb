```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from urllib.parse import urljoin, urlparse

# الرابط الأساسي
base_url = "https://www.meed.com"
page_url = "https://www.meed.com/countries/jordan?location=4906"

os.makedirs("images", exist_ok=True)

data = []

response = requests.get(page_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
article_links = []
for a in soup.find_all("a", href=True):
    href = a["href"].strip()

    # فلترة: نأخذ فقط الروابط التي تشير إلى مقالات داخل نفس الموقع وليست صفحات عامة
    if href.startswith("/") and not href.startswith("/countries") and "?" not in href:
        full_link = urljoin(base_url, href)
        if full_link not in article_links:
            article_links.append(full_link)

print(f"🔗 عدد الروابط داخل الصفحة: {len(article_links)}")
```

```python
from io import BytesIO

for url in article_links:
    try:
        print(f"📄 جارٍ قراءة: {url}")
        article_res = requests.get(url)
        article_soup = BeautifulSoup(article_res.content, "html.parser")

        #  العنوان
        title_tag = article_soup.find("h1")
        title = title_tag.get_text(strip=True) if title_tag else "عنوان غير متوفر"

        #  المقال
        paragraphs = article_soup.find_all("p")
        content = "\n".join(p.get_text(strip=True) for p in paragraphs if p.get_text(strip=True))

        # #  الصورة داخل المقال (إن وجدت)
        # img_tags = article_soup.find_all("img")
        # img_count = 0

        # for img_tag in img_tags:
        #     img_src = img_tag.get("src")
        #     if not img_src:
        #         continue

        #     img_url = urljoin(url, img_src)

        #     try:
        #         img_response = requests.get(img_url, timeout=10)
        #         if img_response.status_code == 200 and len(img_response.content) > 200:
        #             img_ext = os.path.splitext(img_url)[1].split("?")[0] or ".jpg"
        #             img_count += 1
        #             file_path = os.path.join("images", f"{safe_title}_{img_count}{img_ext}")

        #             try:
        #                 image = Image.open(BytesIO(img_response.content))
        #                 image.save(file_path)
        #                 print(f"✅ تم حفظ الصورة (PIL): {file_path}")
        #             except:
        #                 with open(file_path, 'wb') as f:
        #                     f.write(img_response.content)
        #                 print(f"✅ تم حفظ الصورة (RAW): {file_path}")
        #         else:
        #             print(f"⚠️ صورة صغيرة جدًا أو فشلت: {img_url}")
        #     except Exception as e:
        #         print(f"❌ خطأ تحميل الصورة: {img_url} | {e}")

        data.append({
            "العنوان": title,
            "المحتوى": content,
            "الرابط": url
        })

    except Exception as e:
        print(f"❌ خطأ في {url}: {e}")
```
