```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from urllib.parse import urljoin, urlparse

# ÿßŸÑÿ±ÿßÿ®ÿ∑ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿä
base_url = "https://www.meed.com"
page_url = "https://www.meed.com/countries/jordan?location=4906"

os.makedirs("images", exist_ok=True)

data = []

response = requests.get(page_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
article_links = []
for a in soup.find_all("a", href=True):
    href = a["href"].strip()

    # ŸÅŸÑÿ™ÿ±ÿ©: ŸÜÿ£ÿÆÿ∞ ŸÅŸÇÿ∑ ÿßŸÑÿ±Ÿàÿßÿ®ÿ∑ ÿßŸÑÿ™Ÿä ÿ™ÿ¥Ÿäÿ± ÿ•ŸÑŸâ ŸÖŸÇÿßŸÑÿßÿ™ ÿØÿßÿÆŸÑ ŸÜŸÅÿ≥ ÿßŸÑŸÖŸàŸÇÿπ ŸàŸÑŸäÿ≥ÿ™ ÿµŸÅÿ≠ÿßÿ™ ÿπÿßŸÖÿ©
    if href.startswith("/") and not href.startswith("/countries") and "?" not in href:
        full_link = urljoin(base_url, href)
        if full_link not in article_links:
            article_links.append(full_link)

print(f"üîó ÿπÿØÿØ ÿßŸÑÿ±Ÿàÿßÿ®ÿ∑ ÿØÿßÿÆŸÑ ÿßŸÑÿµŸÅÿ≠ÿ©: {len(article_links)}")
```
